```python
for n in range(1, 101):
    if n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    elif n % 3 == 0 & n % 5 == 0:
        print("Fizz Buzz")
    else:
        print(n)
```

```python
movies = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Memento", 9000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]
total_budget = 0
for movie in movies:
    total_budget += movie[1]

average_budget = total_budget / len(movies)
print("Average Budget: ", average_budget)

# Checking movies whose budget is greater then average budget.

above_average = 0
for movie in movies:
    name = movie[0] 
    budget = movie[1]
    
    if budget > average_budget:
        difference = budget - average_budget
        print(f"{name} had higher budget than average budget") 
        above_average += 1
        
print("Total Movies above average: ", above_average)
```
