```python
import math

def minmax(curDepth, nodeIndex, isMaxTurn, scores, targetDepth):
    if curDepth == targetDepth:
        return scores[nodeIndex]


    if isMaxTurn:
        return max(
            minmax(curDepth + 1, nodeIndex * 2, False, scores, targetDepth),
            minmax(curDepth + 1, nodeIndex * 2 + 1, False, scores, targetDepth)
        )
    else:  
        return min(
            minmax(curDepth + 1, nodeIndex * 2, True, scores, targetDepth),
            minmax(curDepth + 1, nodeIndex * 2 + 1, True, scores, targetDepth)
        )

scores = [3, 5, 2, 9, 3, 5, 2, 9]          # Leaf node values
treeDepth = int(math.log(len(scores), 2))   # Depth of the tree

print("The value:", minmax(0, 0, True, scores, treeDepth))
```
