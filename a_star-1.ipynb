```python
class Graph:
    def __init__(self, edges):
        self.edges = edges

    def get_neighbors(self, node):
        return self.edges[node]
    def heuristic(self, node):
        heuristic_values = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        return heuristic_values[node]

    def a_star(self, start, goal):
        open_nodes = [start]
        closed_nodes = []
        g_score = {start: 0}
        parent = {start: start}

        while len(open_nodes) > 0:
            current = None
            for node in open_nodes:
                if current is None or g_score[node] + self.heuristic(node) < g_score[current] + self.heuristic(current):
                    current = node

            if current is None:
                print("Path not exist!")
                return None
            if current == goal:
                path = []
                while parent[current] != current:
                    path.append(current)
                    current = parent[current]
                path.append(start)
                path.reverse()

                print("Path found:", path)
                return path
            for (neighbor, cost) in self.get_neighbors(current):
                if neighbor not in open_nodes and neighbor not in closed_nodes:
                    open_nodes.append(neighbor)
                    parent[neighbor] = current
                    g_score[neighbor] = g_score[current] + cost
                else:
                    if g_score[neighbor] > g_score[current] + cost:
                        g_score[neighbor] = g_score[current] + cost
                        parent[neighbor] = current

                        if neighbor in closed_nodes:
                            closed_nodes.remove(neighbor)
                            open_nodes.append(neighbor)

            open_nodes.remove(current)
            closed_nodes.append(current)

        print("Path does not exist!")
        return None
edges = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)]
}

graph = Graph(edges)
graph.a_star('A', 'D')
```
